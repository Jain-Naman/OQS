```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
pop1_initial1 = np.load("./pop1_initial1.npy")
pop1_initial0 = np.load("./pop1_initial0.npy")
pop1_initial_plus = np.load("./pop1_initial+.npy")
```

```python
pop1_initial1 = np.insert(pop1_initial1, 0, 1.0)
pop1_initial0 = np.insert(pop1_initial0, 0, 0.0)
pop1_initial_plus = np.insert(pop1_initial_plus, 0, 0.5)
```

```python
dt = 0.01
total_time = 0.2
```

```python
time = np.linspace(0, total_time, 21)
time
```

```python
plt.figure(figsize=(8, 6))
plt.plot(time, pop1_initial0, "-o", label="Initial state = |0>")
plt.plot(time, pop1_initial1, "-o", label="Initial state = |1>")
plt.plot(time, pop1_initial_plus, "-o", label="Initial state = |+>")

# plt.ylim(0.47, 0.55)
plt.xlabel("Time")
plt.ylabel("$Tr (\sigma_{+} \sigma_{-} \\rho(t)) $")
plt.title("Population of the excited state over time")
plt.grid()
plt.legend()
plt.show()
```
